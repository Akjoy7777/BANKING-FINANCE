```python
# List the null values
null_Col_list = [features for features in df.columns if df[features].isnull().sum()>0]
print(null_Col_list, end=" ")
print('\n')

# Count of the columns with null as value
count = 0
for loop in null_Col_list:
    count+=1
count
```

```python
# Checking the number and count NaN values
count = 0
for loop in null_Col_list:
    count+=1
    print("[Attribute {}]\n{} : \n[Value count] \n{}, \n[shape] : {}, \n[Null value count] : {}".format(count, loop, df[loop].value_counts(), df[loop].shape, 
                                                                          df[loop].isnull().sum()))
    
    print()
    print("-----------------------------------------------------------------------------")
    print()
```
